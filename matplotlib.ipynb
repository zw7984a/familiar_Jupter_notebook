### Histogrom

```python
np.random.seed(2018)
mu,sigma=100,15
X=mu+sigma*np.random.randn(10000)
#alpha透明度
n, bins, patches = plt.hist(X, 50, normed=1, facecolor='b', alpha=0.5)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
n,bins,patches=plt.hist(weather_2012_final['Temp (C)'],5,facecolor='g',alpha=0.9)

plt.xlabel('Temperture in C')
plt.ylabel('Counts of Days')
plt.title('Histogram of Temperture in 2012')

plt.grid(True)
plt.show()
```

### What is Patches?

```python
#每一个patches代表一个bin
x = np.random.normal(size=100)
n, bins, patches = plt.hist(x)

plt.setp(patches[0],'facecolor','r',alpha=0.5) 
plt.grid()
plt.show()
```

### Pie chart

```python
labels='Cats','Birds','Dogs','Pandas'
sizes=[25,35,30,10]
explode = (0,0.08,0,0)
```

```python
fig1, ax1 = plt.subplots()

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', #数字的位置
        shadow=False, startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# Show
plt.show()
```

### About Sub-plots

```python
# plot a line, implicitly creating a subplot(111)
plt.plot([1,2,3])
# now create a subplot which represents the top plot of a grid
# with 2 rows and 1 column. Since this subplot will overlap the
# first, the plot (and its axes) previously created, will be removed
plt.subplot(211) #表示两行一列第一张图
plt.plot(range(12))
plt.subplot(212, facecolor='y') # creates 2nd subplot with yellow background
```
