### graphviz

```python
import graphviz
```

```python
import sklearn.datasets as datasets
import pandas as pd
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

numpy：numpy定义了python进行矩阵数值计算的基础

np.add(A,B) 数组对应元素加法

np.subtract(A,B) 数组对应元素减法

np.dot(C,D) # 矩阵乘法

np.matmul(C,D) ## C@D 矩阵乘法

np.multiply(C,D) ## C*D 数组对应元素相乘



重点看看 np.dot 和 np.matmul的区别：

都是矩阵乘法
np.matmul中，禁止矩阵与标量的乘法
np.matmul中，在矢量乘矢量的内积运算中，matmul与dot没有差别
np.matmul中，多维的矩阵，将前n-2维视为后2维的元素后，进行乘法操作

```python
import numpy as np
```

```python
A = np.array([1,2,3]) #array变成数组
B = np.array([4,3,0])
print('A+B = ',np.add(A,B))
print('A-B = ',np.subtract(A,B))
```

```python
#元素与元素相乘
A = np.array([1,2,3])
print('(A,A)=',np.multiply(A,A))
print('(A,A)=', A*A)
```

```python
C = np.array([[1,2,3],
            [1,1,1]])
D = np.array([[4,3,9,9],
            [1,1,2,2,],
            [4,2,1,1]])
```

```python
C.shape # C的纬度
D.shape # D的纬度
```

```python
print('np.dot(C,D) = ',np.dot(C,D))
print('np.matmul(C,D) = ',np.matmul(C,D))
print('        C@D    = ',np.matmul(C,D))

print('np.multiply(D,D)=',np.multiply(D,D))
print('          D*D   = ', D*D)
```
